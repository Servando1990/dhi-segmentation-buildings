```python
import sys
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
fig,axs=plt.subplots(2, 3,figsize=(9,6))
axs[0,0].set_title('Adv Loss Train')
axs[1,0].set_title('Adv Loss Val')
axs[0,1].set_title('Gene Target Loss Train')
axs[1,1].set_title('Gene Target Loss Val')
axs[1,2].set_title('Adv accuracy Val')

LOSS_FILE='LOSSES_COMPARISON_ADDA/'

if not os.path.exists(LOSS_FILE):
            os.makedirs(LOSS_FILE)
    

FILE_NAME=LOSS_FILE+'ghana_adda_test'
f1='MODEL_ADDA_2/STORE_LEARNING_ADDA/'

with open(f1+'avg_loss_train_d.txt', "r",) as ins:
    loss_tr_d = []
    for line in ins.readlines():
        loss_tr_d.append(float(line))
            
loss_tr_d=np.asarray(loss_tr_d)
epoch = np.arange(0,len(loss_tr_d),1)

with open(f1+'loss_verif_d.txt', "r",) as ins:
    loss_val_d = []
    for line in ins.readlines():
        loss_val_d.append(float(line))
            
loss_val_d=np.asarray(loss_val_d)
epoch_val = np.arange(0,len(loss_val_d),1)

with open(f1+'avg_loss_train_g.txt', "r",) as ins:
    loss_tr_g = []
    for line in ins.readlines():
        loss_tr_g.append(float(line))
            
loss_tr_g=np.asarray(loss_tr_g)

with open(f1+'loss_verif_g.txt', "r",) as ins:
    loss_val_g = []
    for line in ins.readlines():
        loss_val_g.append(float(line))
            
loss_val_g=np.asarray(loss_val_g)


with open(f1+'acc_verif_d.txt', "r",) as ins:
    acc_verif_d = []
    for line in ins.readlines():
        acc_verif_d.append(float(line))
            
acc_verif_d=np.asarray(acc_verif_d)



axs[0,0].plot(epoch, loss_tr_d, 'r-') 
axs[1,0].plot(epoch_val, loss_val_d, 'g-')
axs[0,1].plot(epoch, loss_tr_g,
              'r-') 
axs[1,1].plot(epoch_val, loss_val_g, 'g-')
axs[1,2].plot(epoch_val, acc_verif_d*100, 'g-') 
axs[0,2].axis('off')
fig.savefig(FILE_NAME+'_stats_adda.png')   
```
