```python
import sys
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python

LOSS_FILE='LOSSES_COMPARISON/'

if not os.path.exists(LOSS_FILE):
            os.makedirs(LOSS_FILE)
    

FILE_NAME=LOSS_FILE+'ghana_adda_test'
# FILE_NAME=LOSS_FILE+'spacenet_test'

step=5

# f1='MODEL_ADDA_TEST/STORE_LEARNING/'
f1='MODEL_ADDA_2/STORE_LEARNING/'
# f1='MODEL_TEST_SPACENET/STORE_LEARNING/'

leg=['MODEL TEST GHANA']
# leg=['MODEL TEST SPACENET']
# leg=['MODEL TEST SPACENET','dummy']


fig,axs=plt.subplots(1, 4,figsize=(15,3))



#TRAIN


with open(f1+'avg_loss_train.txt', "r",) as ins:
    t1 = []
    for line in ins.readlines():
        t1.append(float(line))
            
t1=np.asarray(t1)
ite1 = np.arange(0,len(t1),1)


# with open(f2+'avg_loss_train.txt', "r") as ins:
#     t2 = []
#     for line in ins.readlines():
#         t2.append(float(line))
# t2=np.asarray(t2)
# ite2 = np.arange(0,len(t2),1)


# axs[0].set_ylim(0.3,0.5)
axs[0].plot(ite1,t1,'b', label=leg[0])
# axs[0].plot(ite2,t2,'g', label=leg[1])
axs[0].set_title('TRAINING')
axs[0].set(ylabel='LOSS', xlabel='iterations')
axs[0].legend(loc='upper right',fontsize='small') 


#VERIF LOSS
with open(f1+'loss_verif.txt', "r") as ins:
    t1 = []
    for line in ins.readlines():
        t1.append(float(line))
t1=np.asarray(t1)
ite1 = np.arange(0,len(t1),1)
print(leg[0])
print("Minimum verif loss :%.4f, Model number %d"%(np.amin(t1),np.argmin(t1)))



# with open(f2+'loss_verif.txt', "r") as ins:
#     t2 = []
#     for line in ins.readlines():
#         t2.append(float(line)) 
# t2=np.asarray(t2)
# ite2 = np.arange(0,len(t2),1)

# print(leg[1])
# print("Minimum verif loss :%.4f, Model number %d"%(np.amin(t2),np.argmin(t2)))

axs[1].plot(ite1,t1,'b', label=leg[0])
# axs[1].plot(ite2,t2,'g', label=leg[1])
axs[1].set_title('VERIFICATION')
axs[1].set(ylabel='LOSS', xlabel='epochs')
axs[1].legend(loc='upper right',fontsize='small') 

#ERROR RATE

with open(f1+'error_verif.txt', "r") as ins:
    t1 = []
    for line in ins.readlines():
        t1.append(float(line))
t1=np.asarray(t1)
ite1 = np.arange(0,len(t1),1)

print(leg[0])
print("Minimum error rate:%.4f, Model number %d"%(np.amin(t1),np.argmin(t1)))

# with open(f2+'error_verif.txt', "r") as ins:
#     t2 = []
#     for line in ins.readlines():
#         t2.append(float(line))
# t2=np.asarray(t2)
# ite2 = np.arange(0,len(t2),1)

# print(leg[1])
# print("Minimum error rate:%.4f, Model number %d"%(np.amin(t2),np.argmin(t2)))


# axs[2].set_ylim(4,5)
axs[2].plot(ite1,t1,'b', label=leg[0])  
# axs[2].plot(ite2,t2,'g', label=leg[1])
axs[2].set_title('VERIFICATION')
axs[2].set(ylabel='Error rate', xlabel='epochs')
axs[2].legend(loc='upper right',fontsize='small')  




## F1 SCORE


with open(f1+'f1_iou_verif.txt', "r") as ins:
    t1 = []
    for line in ins.readlines():
        t1.append(float(line))
t1=np.asarray(t1)
ite1 = np.arange(step,(len(t1)+1)*step,step)

# with open(f2+'f1_iou_verif.txt', "r") as ins:
#     t2 = []
#     for line in ins.readlines():
#         t2.append(float(line))
# t2=np.asarray(t2)
# ite2 = np.arange(step,(len(t2)+1)*step,step)


axs[3].plot(ite1,t1,'b', label=leg[0])  
# axs[3].plot(ite2,t2,'g', label=leg[1]) 
axs[3].set_title('VERIFICATION')
axs[3].set(ylabel='f1 score IOU', xlabel='epochs')
axs[3].legend(loc='lower right',fontsize='small')

fig.savefig(FILE_NAME+'_stats_classif_target.png')   


plt.show()
```
