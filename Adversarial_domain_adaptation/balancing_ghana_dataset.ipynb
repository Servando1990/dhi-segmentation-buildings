```python
import sys
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
import logging
from torch.utils.data import DataLoader
import torch
import torch.nn as nn
from torchvision import transforms
import matplotlib.pyplot as plt


sys.path.append(os.path.abspath('../'))

from predict_and_evaluate import *
from Data_Handle.image_utils import *
from Data_Handle.dataset_generator import Dataset_sat
```

```python
data_root_ghana_balanced="../../2_DATA_GHANA_BALANCED/DATASET/128_x_128_8_pansh/"

if not os.path.exists(data_root_ghana_balanced):
            os.makedirs(data_root_ghana_balanced)
```

```python
data_root_ghana = "../../2_DATA_GHANA/DATASET/128_x_128_8_pansh/"

#TRAINING
path_training=os.path.join(data_root_ghana,"TRAINING/")
if not os.path.exists(os.path.join(data_root_ghana_balanced,"TRAINING/INPUT/")):
            os.makedirs(os.path.join(data_root_ghana_balanced,"TRAINING/INPUT/"))
if not os.path.exists(os.path.join(data_root_ghana_balanced,"TRAINING/OUTPUT/")):
            os.makedirs(os.path.join(data_root_ghana_balanced,"TRAINING/OUTPUT/"))
#CREATE  BALANCED training/valid/verif and input + output
generator_tr=Dataset_sat.from_root_folder(path_training,2,max_data_size=None)
loader_tr=DataLoader(generator_tr, batch_size=1,shuffle=False, num_workers=1)

count=0
for batch,sample in enumerate(loader_tr):
    X=np.squeeze(np.asarray(sample["input"]))
    Y=np.asarray(sample["groundtruth"])
    Y_flat=np.squeeze(np.argmax(Y,3))
    if np.sum(Y_flat)==0:
        continue
    else:
        write_data_h5(os.path.join(data_root_ghana_balanced,"TRAINING/INPUT/input_"+str(count)+".h5"),X)
        write_data_h5(os.path.join(data_root_ghana_balanced,"TRAINING/OUTPUT/output_"+str(count)+".h5"),Y_flat)
        count+=1
    
#VALIDATION
path_validation=os.path.join(data_root_ghana,"VALIDATION/")
if not os.path.exists(os.path.join(data_root_ghana_balanced,"VALIDATION/INPUT/")):
            os.makedirs(os.path.join(data_root_ghana_balanced,"VALIDATION/INPUT/"))
if not os.path.exists(os.path.join(data_root_ghana_balanced,"VALIDATION/OUTPUT/")):
            os.makedirs(os.path.join(data_root_ghana_balanced,"VALIDATION/OUTPUT/"))
#CREATE  BALANCED training/valid/verif and input + output
generator_val=Dataset_sat.from_root_folder(path_validation,2,max_data_size=None)
loader_val=DataLoader(generator_val, batch_size=1,shuffle=False, num_workers=1)


for batch,sample in enumerate(loader_val):
    X=np.squeeze(np.asarray(sample["input"]))
    Y=np.asarray(sample["groundtruth"])
    Y_flat=np.squeeze(np.argmax(Y,3))
    if np.sum(Y_flat)==0:
        continue
    else:
        write_data_h5(os.path.join(data_root_ghana_balanced,"VALIDATION/INPUT/input_"+str(count)+".h5"),X)
        write_data_h5(os.path.join(data_root_ghana_balanced,"VALIDATION/OUTPUT/output_"+str(count)+".h5"),Y_flat)
        count+=1
#TEST
path_test=os.path.join(data_root_ghana,"TEST/")
if not os.path.exists(os.path.join(data_root_ghana_balanced,"TEST/INPUT/")):
            os.makedirs(os.path.join(data_root_ghana_balanced,"TEST/INPUT/"))
if not os.path.exists(os.path.join(data_root_ghana_balanced,"TEST/OUTPUT/")):
            os.makedirs(os.path.join(data_root_ghana_balanced,"TEST/OUTPUT/"))
#CREATE  BALANCED training/valid/verif and input + output
generator_test=Dataset_sat.from_root_folder(path_test,2,max_data_size=None)
loader_test=DataLoader(generator_test, batch_size=1,shuffle=False, num_workers=1)


for batch,sample in enumerate(loader_test):
    X=np.squeeze(np.asarray(sample["input"]))
    Y=np.asarray(sample["groundtruth"])
    Y_flat=np.squeeze(np.argmax(Y,3))
    if np.sum(Y_flat)==0:
        continue
    else:
        write_data_h5(os.path.join(data_root_ghana_balanced,"TEST/INPUT/input_"+str(count)+".h5"),X)
        write_data_h5(os.path.join(data_root_ghana_balanced,"TEST/OUTPUT/output_"+str(count)+".h5"),Y_flat)
        count+=1
                                                 
    
```
